# Quora Question Pairs

The aim of this project is to identify question pairs that are duplicates. The raw data given provides us with already identified question pairs, along with a flag called "is_duplicate", that highlights if the question pair is a duplicate or not.

Since there are no features for a machine learning algorithm to work its magic on, we have to first identify possible features that can be extracted from these question pairs.

1. Number of important words that are present in both questions in a question pair
The first most obvious way of detecting duplicate question pairs is when the exact same words are used. Most usually, the higher the number of words present in both questions in a question pair, the more similar the questions.

2. Whether the type of interoggative word used (either Y/N)
A 'why' question is very different from a 'how' question. 
i.e. Why is batman so strong? vs How is batman so strong?

3. Total tf-idf score of a sentence
Pronouns that possess large tf-idf values may result in a similar, but still NOT duplicate question.
i.e. What is the most common name? vs What is the most common name in Singapore?
Similar, yet very different. 

4. Total vector distance between each sentence
Using word to vec, words can not be projected onto vector space. Ideally, completely duplicate questions should have a total vector distance of 0. So, the closer the total vector distances is to 0, the higher the likelihood of the question pair being a duplicate.


So first, the data is loaded. Questions from each question pair are loaded into separate lists.

```python
import pandas as pd
train = pd.read_csv('/Users/matthewyeozhiwei/repos/KaggleQuora/train.csv')
questions1 = train['question1'].tolist()
questions2 = train['question2'].tolist()
```

## Num. of Impt Words Present in both Questions aka NumWords

To get the NumWords, we will need to do preprocessing. This consists of normalizing, stop-word removal followed by stemming. This will remove stopwords and remove grammar from causing any disruptions in the analysis. 

```python
import string
sp = string.punctuation

def normalize(questions):
    questions = list(map(lambda t: ''.join(["" if c.isdigit() else c for c in str(t)]), questions))
    questions = list(map(lambda t: ''.join(["" if c in sp else c for c in str(t)]), questions))
    questions = list(map(str.lower, questions))
    return questions
```

```python
questions1n = normalize(questions1)
questions2n = normalize(questions2)
```

To better visualize why stopwords are noisy, cumulative and frequency graphs can be plotted. But, first, the question lists have to be converted into term frequency.

```python
def to_TF(questions):
    import pandas as pd
    import nltk
    questions = list(map(lambda t: nltk.regexp_tokenize(t, r'\S+'), questions))
    questions = [w for l in questions for w in l]

    ## Compute the frequency distribution of the words as a dictionary
    fdist = nltk.FreqDist(questions) 
    ## Convert the dictionary to a dataframe contaning the words and
    ## counts indexed by the words, and then take the transpose.
    count_frame = pd.DataFrame(fdist, index =[0]).T
    count_frame.columns = ['Count']
    return(count_frame.sort_values('Count', ascending = False))
```

```python
def wf_bar(wf):
    import matplotlib.pyplot as plt
    ## Barplot of the most fequent words.   
    fig = plt.figure(figsize=(12, 9))
    ax = fig.gca()    
    wf['Count'][:60].plot(kind = 'bar', ax = ax)
    ax.set_title('Frequency of the most common words')
    ax.set_ylabel('Frequency of word')
    ax.set_xlabel('Word')
    plt.show()
    return 'Done'


def plot_cfd(wf):
    import matplotlib.pyplot as plt
    ## Compute the relative cumulative frequency of the words in 
    ## descending order of frequency and add the dataframe.   
    word_count = float(wf['Count'].sum(axis = 0))   
    wf['Cum'] = wf['Count'].cumsum(axis = 0)
    wf['Cum'] = wf['Cum'].divide(word_count)
    
    ## Barplot the cumulative frequency for the most frequent words.   
    fig = plt.figure(figsize=(12, 9))
    ax = fig.gca()    
    wf['Cum'][:60].plot(kind = 'bar', ax = ax)
    ax.set_title('Cumulative fraction of total words vs. words')
    ax.set_ylabel('Cumulative fraction')
    ax.set_xlabel('Word')
    plt.show()
    return 'Done'
```

```python
wf1 = to_TF(questions1n)
wf_bar(wf1)
plot_cfd(wf1)
```

```python
wf2 = to_TF(questions2n)
wf_bar(wf2)
plot_cfd(wf2)
```

```python
stop_words = pd.read_csv('/Users/matthewyeozhiwei/repos/KaggleQuora/stopwords.csv')
stop_words = [w for w in stop_words.words if w in stop_words.words.unique() ]

def rmstop(questions):
    temp = [question.split() for question in questions]
    questions = [' '.join([word for word in question if word not in set(stop_words)]) for question in temp]
    return questions
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
def stem(questions):
    temp = [question.split() for question in questions] 
    temp = map(lambda t: [porter_stemmer.stem(w) for w in t], temp)
    questions = [' '.join(question) for question in temp] 
    return questions
```

```python
questions1n = rmstop(questions1n)
questions2n = rmstop(questions2n)
questions1n = stem(questions1n)
questions2n = stem(questions2n)
```

```python
wf1 = to_TF(questions1n)
wf_bar(wf1)
plot_cfd(wf1)
```

```python
wf2 = to_TF(questions2n)
wf_bar(wf2)
plot_cfd(wf2)
```
